```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
daimond=pd.read_csv('/content/Diamonds Prices2022.csv')
daimond.head(6)
```

```python
plt.scatter(data=daimond, x='depth',y='x')
plt.xlabel('Displacement')
plt.ylabel('Combined daimond Efficiency')
```

```python
sns.regplot(data=daimond, x='depth',y='x')
plt.xlabel('Displacement')
plt.ylabel('Combined daimond Efficiency')
```

```python
sns.regplot(data=daimond, x='table',y='depth')
```

```python
sns.regplot(data=daimond, x='table',y='depth',x_jitter=0.3)
```

```python
sns.regplot(data=daimond, x='table',y='depth',x_jitter=0.3,
            scatter_kws={'alpha': 1/20});
```

```python
plt.hist2d(data=daimond, x='table',y='depth')
plt.colorbar()
plt.xlabel('Displacement')
plt.ylabel('Combined daimond Efficiency')
```

```python
plt.hist2d(data=daimond, x='table',y='depth',cmin=0.5);
plt.colorbar()
plt.xlabel('Displacement')
plt.ylabel('Combined daimond Efficiency')
```

```python
plt.hist2d(data=daimond, x='table',y='depth',cmin=0.5,cmap='viridis_r');
plt.colorbar()
plt.xlabel('Displacement')
plt.ylabel('Combined daimond Efficiency')
```

```python
base_color = sns.color_palette()[0]
sns.boxplot(data=daimond, x='cut', y='table',color=base_color)
plt.xticks(rotation=15)
```

```python
daimond['color_type'] = daimond['color'].apply(lambda x: x.split()[0])
```

```python
daimond['color'].head()
```

```python
ct_counts = daimond.groupby(['cut','color_type']).size()
ct_counts = ct_counts.reset_index(name='count')
ct_counts.pivot(index='cut',columns='color_type',values='count')
```

```python
sns.countplot(data=daimond,x='cut',hue='color_type')
plt.xticks(rotation=15)
```

```python
bins = np.arange(12,58+2,2)
g = sns.FacetGrid(data=daimond,col='cut',col_wrap=3)
g.map(plt.hist,'table',bins=bins);
```

```python
base_color = sns.color_palette()[7]
sns.barplot(data=daimond,x='cut',y='table',color=base_color)
plt.xticks(rotation=15)
plt.ylabel('Average daimond Efficiency')
```

```python
sns.pointplot(data = daimond, x = 'cut', y = 'table')
plt.ylabel('Average daimond Efficiency')
plt.xticks(rotation=15)
```
